# Adding coulmn total to existing dataset

```python
#calculating column totals

sum_tot=df[['sepal length','sepal width','petal length','petal width','Sum Row']].sum()
sum_tot
```

```python
#transposing the sums and forming a dataset

T_sum= pd.DataFrame(data=sum_tot).T
T_sum
```

```python
#rearraging columns to replicate existing dataframe

T_sum = T_sum.reindex(columns=df.columns)
T_sum
```

```python
#merging it back with dataframe

Row_Total = df.append(T_sum,ignore_index=True)
Row_Total
```

# Pivot Tables

```python
pd.pivot_table(df, index= 'class')  #same as groupby
```

```python
#cross tab where index=row #replacing NA with 0
import numpy as np
pd.pivot_table(df, values='sepal width', index= 'sepal length',columns='class', aggfunc = np.sum,fill_value=0).head(10)
```

```python
pd.pivot_table(df, values=['sepal width', 'petal width'], index= 'sepal length',
               columns='class', aggfunc = np.sum, fill_value=0).head()
```

```python
pd.pivot_table(df, values=['sepal width', 'petal width'], index= ['sepal length', 'petal length'],
              columns='class', aggfunc = np.sum, fill_value=0).head()
```

```python
#margins adds row and column total

pd.pivot_table(df, values=['sepal width', 'petal width'], index= 'sepal length',columns='class', 
               aggfunc = {'sepal width': np.sum, 'petal width': np.mean}, fill_value=0, margins=True)
```

```python
pivoted=pd.pivot_table(df, values=['sepal width', 'petal width'], index= 'sepal length',columns='class', 
               aggfunc = {'sepal width': np.sum, 'petal width': np.mean}, fill_value=0, margins=True)
```
